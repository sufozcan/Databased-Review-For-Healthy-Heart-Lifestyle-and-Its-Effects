```python
import pandas as pd #importing libraries
import numpy as np
```

```python
dirty_data = pd.read_excel('Cleandata1.xlsx') #importing step1 clean dataset
```

```python
dirty_data = dirty_data.rename({'Execise Duration': 'Exercise Duration', 'Body Mass Index':'BMI'},axis= "columns")
dirty_data.columns # handling with previous problems caused by our carelessness
```

```python
dirty_data.head(15) # Checking whether importing wrong dataset
```

```python
dirty_data.drop('Unnamed: 0', axis=1, inplace=True)
print(dirty_data) # Dropping unnecessary column which comes from colab
```

```python
dirty_data.head() # Checking again and we saw that gender and smoking status sections are lower case while we have written title format in previous step so we handle that problem in below
```

```python
dirty_data.loc[dirty_data['Gender'].isin(['female', 'f']), 'Gender'] = 'Female'
dirty_data.loc[dirty_data['Gender'].isin(['male', 'm']), 'Gender'] = 'Male'
```

```python
dirty_data.loc[dirty_data['Smoking Status'].isin(['non-smoker']), 'Smoking Status'] = 'Non-smoker'
dirty_data.loc[dirty_data['Smoking Status'].isin(['smoker']), 'Smoking Status'] = 'Smoker'
```

```python
dirty_data.head() # Last controlling before passing through to numerical variables cleanings
```

```python
dirty_data['Exercise Duration'].values # starting with exercise duration section by looking at the values
```

```python
dirty_data['Exercise Duration'] = dirty_data['Exercise Duration'].replace({
    '90,0': '90',
    '150,0': '150',
    '120,0': '120',
    '60,0': '60',
    '-120,0': '-120',
    '-60,0': '-60',
    '1500,0': '150'
})
dirty_data['Exercise Duration'] = dirty_data['Exercise Duration'].astype(float).astype('Int64')  # In here, we wanted to handling with integer values because of its simplicity but we will change all numerical values to float at the end.
```

```python
print(dirty_data['Exercise Duration'].value_counts())
```

```python
dirty_data['Exercise Duration'].loc[dirty_data['Exercise Duration'] == int('-120')] = int('120') # Here we change - values to their absoulute values by directly changing
dirty_data['Exercise Duration'].loc[dirty_data['Exercise Duration'] == int('-60')] = int('60')
```

```python
print(dirty_data['Exercise Duration'].value_counts()) # Checking before passing another numerical cleaning
```

```python
dirty_data.head() # looking for cleaning
```

```python
dirty_data['Age'] = dirty_data['Age'].astype(float).astype('Int64') # here we would like to change data type integer to float because of the checklist's item
```

```python
dirty_data['Glucose Level'] = dirty_data['Glucose Level'].astype(float).astype('Int64')
```

```python
dirty_data.head() # Looking for cleaning and found BMI
```

```python
print(dirty_data['BMI'].value_counts()) # Looking BMI variables
```

```python
dirty_data['BMI'].values # Looking more detailed for values
```

```python
dirty_data['BMI'].loc[dirty_data['BMI'] == '27,6'] = '27.6' # here we would like to deal with one by one because of it looked like more easy
dirty_data['BMI'].loc[dirty_data['BMI'] == '24,6'] = '24.6'
dirty_data['BMI'].loc[dirty_data['BMI'] == '28,4'] = '28.4'
dirty_data['BMI'].loc[dirty_data['BMI'] == '23,1'] = '23.1'
dirty_data['BMI'].loc[dirty_data['BMI'] == '22,4'] = '22.4'
```

```python
print(dirty_data['BMI'].value_counts()) # Checking values
```
