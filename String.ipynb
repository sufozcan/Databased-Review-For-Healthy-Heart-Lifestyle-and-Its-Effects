```python
dirty_data.columns = dirty_data.columns.str.title() # Cleaning column names
dirty_data.columns
```

```python
dirty_data = dirty_data.rename({'Id':'ID', '##Age' : 'Age', '"Gender"' : 'Gender', '"Blood_Pressure"' : 'Blood Pressure',
'"Cholesterol_Level"': 'Cholesterol Level', '"Family Historu"': 'Family History', '"Smoking"': 'Smoking Status', '"Physical_Activity"' :
                                'Physical Activity', '"Body_Mass_Index"': 'Body Mass Index', '"Glucose_Level"' : 'Glucose Level',
                                'Year///' : 'Year', '"Heart_Disease"' : 'Heart Disease', '###Exercise_Duration': 'Execise Duration'},axis= "columns")
# Renaming column names what we wanted to look like
dirty_data.columns
```

```python
dirty_data.duplicated().sum() # Checking duplicates
```

```python
dirty_data = dirty_data.drop_duplicates() # Cleaning them and check again
dirty_data.duplicated().sum()
```

```python
print(dirty_data['Gender'].value_counts()) # Passing through to variables and starting with gender section
```

```python
dirty_data['Gender'] = dirty_data['Gender'].str.lower() # Cleaning steps by changing all values to lower case and after that changing the form we wanted
```

```python
dirty_data.loc[dirty_data['Gender'].isin(['female', 'f']), 'Gender'] = 'Female'
dirty_data.loc[dirty_data['Gender'].isin(['male', 'm']), 'Gender'] = 'Male'
print(dirty_data['Gender'].value_counts())
```

```python
dirty_data['Smoking Status'] = dirty_data['Smoking Status'].str.lower() # Here we wanted to again change all values to lower case and afterly change the form we wanted
```

```python
dirty_data.loc[dirty_data['Smoking Status'].isin(['non-smoker']), 'Smoking Status'] = 'Non-smoker'
dirty_data.loc[dirty_data['Smoking Status'].isin(['smoker']), 'Smoking Status'] = 'Smoker'
dirty_data.loc[dirty_data['Smoking Status'].isin(['non_smoker']), 'Smoking Status'] = 'Non-smoker'
```

```python
print(dirty_data['Smoking Status'].value_counts()) # Controlling the final
```

```python
dirty_data['Physical Activity'] = dirty_data['Physical Activity'].str.strip() # This section's cleaning steps starts with handling white space problem
print(dirty_data['Physical Activity'].value_counts()) # In here we run the code after cleaning because of that action here it looks like clean
```

```python
dirty_data['Physical Activity'] = dirty_data['Physical Activity'].str.lower() #we wanted to firtsly lower all of variables and then change the form we wanted
```

```python
dirty_data.loc[dirty_data['Physical Activity'].isin(['moderate']), 'Physical Activity'] = 'Moderate'
dirty_data.loc[dirty_data['Physical Activity'].isin(['low']), 'Physical Activity'] = 'Low'
dirty_data.loc[dirty_data['Physical Activity'].isin(['high']), 'Physical Activity'] = 'High'
dirty_data.loc[dirty_data['Physical Activity'].isin(['m']), 'Physical Activity'] = 'Moderate'
dirty_data.loc[dirty_data['Physical Activity'].isin(['l']), 'Physical Activity'] = 'Low'
dirty_data.loc[dirty_data['Physical Activity'].isin(['h']), 'Physical Activity'] = 'High'
```

```python
print(dirty_data['Physical Activity'].value_counts()) # Now it is clean
```

```python
print(dirty_data['Heart Disease'].value_counts()) # Passing through to hearth disease section's cleaning
```

```python
dirty_data['Heart Disease'] = dirty_data['Heart Disease'].str.strip() # Handling white space problem and checking
print(dirty_data['Heart Disease'].value_counts())
```

```python
dirty_data['Heart Disease'] = dirty_data['Heart Disease'].str.lower() # Again we wanted to start with lower all of variables and change them after
```

```python
dirty_data.loc[dirty_data['Heart Disease'].isin(['yes']), 'Heart Disease'] = 'Yes'
dirty_data.loc[dirty_data['Heart Disease'].isin(['no']), 'Heart Disease'] = 'No'
dirty_data.loc[dirty_data['Heart Disease'].isin(['y']), 'Heart Disease'] = 'Yes'
dirty_data.loc[dirty_data['Heart Disease'].isin(['n']), 'Heart Disease'] = 'No'
```
